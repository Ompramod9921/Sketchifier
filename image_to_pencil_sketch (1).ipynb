## Convert any image to a pencil sketch â€” with just 2 lines of code ! 

*using sketchify - It is a simple sketching package in python which converts images into drawings*

*sketchify depends on opencv-contrib-python*

```python
#First, install the package sketchify using pip
pip install sketchify
```

```python
# importing sketch function from sketchify module
from sketchify import sketch
```

```python
# open original image
from IPython.display import Image
Image("/content/elephant.jpeg")
```

```python
#We only need two lines of code to convert any image to a pencil sketch. 
#Sketchify saves the image as a png file, so the Sketchified image will be saved in the folder specified (path/to/save) as image.png.
sketch.normalsketch('/content/elephant.jpeg', '/content/', 'sketch.jpeg')
```

```python
#open picture after pencil sketch effect
from IPython.display import Image
Image("/content/sketch.png")
```

*In this way can convert any photo into a realistic pencil sketch.*
